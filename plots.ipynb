```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_functions(mu:float):
    
    lambda_values = np.arange(0, 2*mu, 0.0001)
    
    functions = [   
                    (1, lambda mu, l: 1/(2*mu - l)),
                    (2, lambda mu, l: 1/(mu -  l/2)),
                    (3, lambda mu, l: (2*mu)/( (2*mu*mu) -  (l*l/2)))
                ]
    
    fig, ax = plt.subplots()
    ax.set_xlabel("$\lambda$")
    ax.set_ylabel("$\overline{R}$")
    ax.set_ylim(0,6)
    
    for (system, function) in functions:
        
        response_times = function(mu,lambda_values)
        
        ax.plot(lambda_values, response_times, label="$\overline{R}"+f"_{system}$")
        
    ax.legend()
plot_functions(1)
```
