```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_functions(mu:float):
    
    lambda_values = np.arange(0, 2*mu, 0.0001)
    
    functions = [   
                    (1,"b", lambda mu, l: 1/(2*mu - l)),
                    (2,"m", lambda mu, l: 1/(mu -  l/2)),
                    (3,"c", lambda mu, l: (2*mu)/( (2*mu*mu) -  (l*l/2)))
                ]
    
    fig, ax = plt.subplots(figsize=(10,8))
    ax.set_xlabel("arrival rate ($\lambda$)")
    ax.set_ylabel("average response time ($\overline{R})$")
    ax.set_ylim(0,8)
    
    ax.axvspan(0, .8, facecolor='green', alpha=0.25, label="Low load")
    ax.axvspan(.8, 1.4, facecolor='yellow', alpha=0.25, label="Medium load")
    ax.axvspan(1.4, 2, facecolor='red', alpha=0.25, label="High load")
    
    for (system, color, function) in functions:
        
        response_times = function(mu,lambda_values)
        
        ax.plot(lambda_values, response_times, label="$\overline{R}"+f"_{system}$", color=color)
    
    fig.legend(loc='upper left', bbox_to_anchor=(0.18, 0.85))
plot_functions(1)
```
